```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error
```

```python
data = pd.read_csv("/content/IMDb Movies India.csv" ,encoding='latin-1')
data
```

```python
#data['Actors'] = data['Actor 1'] +" , " +  data['Actor 2'] + " , " + data['Actor 3']
```

```python
data.drop(columns=['Name' , 'Year' , 'Duration', 'Votes'],inplace=True)
```

```python
data
```

```python
data.dropna(inplace = True)
```

```python
Y = data['Rating']
features = ['Genre' , 'Director' , 'Actor 1' , 'Actor 2' , 'Actor 3']
X = data[features]
```

```python
one_hot_encoded_data = pd.get_dummies(X, columns = ['Genre', 'Director' , 'Actor 1' , 'Actor 2' , 'Actor 3'])
```

```python
one_hot_encoded_data.head(5)
```

```python
X_train, X_test, y_train, y_test = train_test_split(one_hot_encoded_data, Y, test_size = 0.2 , random_state=1)
```

```python
model = DecisionTreeRegressor()
model.fit(X_train, y_train)
preds_val = model.predict(X_test)
mae = mean_absolute_error(y_test, preds_val)
```

```python
print("The model is trained with mean squared error of " , mae)
```

```python
dataF = {'Actual': y_test,
        'Predicted': preds_val}
pd.DataFrame(dataF)
```
